### Import all modules and libraries need.

```python
%matplotlib inline

import numpy as np
import scipy
import matplotlib.pyplot as plt

from scipy.io.wavfile import read
from scipy.signal import hann
```

### Declare various functions needed for each step in the process.

```python
def getwavedata(file):
    '''Imports sound data from a wav file via scipy and outputs a numpy.ndarray.'''
    # item 0 is the header info and item 1 is the audio data itself
    return scipy.io.wavfile.read(file)
```

```python
# function to take in data and file name and output simple plotted graph files
def makegraph(sound, timeRange, sample_rate, title=''):
    plt.figure()
    plt.plot(timeRange, sound)
    plt.ylabel("Amplitude")
    plt.xlabel("Time (seconds)")
    plt.title(title)
    plt.suptitle("Sampled at {}Hz".format(sample_rate))
```

```python
def FFTwithHann(snd, rate):
    ''' Function will take in a numpy array from sound data 
        and output the frequency and power data as a tuple. '''
     # get shape from sound's numpy array
    shape = snd.shape[0]
    # create and apply the Hanning Window filter to audio data
    window = hann(shape)
    snd = snd * window
    # calculate the power and frequencies in audio sample
    power = 20*np.log10(np.abs(np.fft.rfft(snd)))
    freq = np.linspace(0, rate/2, len(power))
    # output a tuple of numpy arrays for downstream processing and analysis
    return (freq, power)
```

```python
# function to plot the frequency spectrum on a graph with appropriate labels
def plotFreqSpect(freqSpect): 
    plt.title('Frequency Spectrum')
    plt.xlabel('Hz')
    plt.ylabel('dB')
    plt.plot(freqSpect[0], freqSpect[1]) # (freq, power)
```

```python
# TO DO: work in 0.05s overlap
def processAudio(snd):
    ''' Function will process an audio sample in 0.25s intervals to output frequency spectrum data. '''
    rate = snd[0]
    step = 1000 # bins in array to sample (~0.25s)
    max_bin = 2000 #snd.shape[0]-1000
    start_bin = 0
    freqSpect = []
    # step over the numpy array and process the audio
    while start_bin < max_bin:
        snippet = snd[1][start_bin:(start_bin+1000)]
        start_bin += step
        freqSpect.append(FFTwithHann(snippet, rate))
    return freqSpect
```
