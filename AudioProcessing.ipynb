```python
%matplotlib inline
import numpy as np
import scipy
from scipy.io.wavfile import read
from scipy.signal import hann, gaussian
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
# import sound data from wav files via scipy into numpy.ndarray
def getwavedata(file):
    # item 0 is the header info and item 1 is the audio data itself
    return scipy.io.wavfile.read(file)

# process both audio wav files
sound4khz = getwavedata('/home/kak9699/Documents/code/RFCx/SoundFiltering/4khz-5m-nt.wav')
# sound8khz = getwavedata('/home/kak9699/Documents/code/RFCx/SoundFiltering/8khz-5m-nt.wav')

# print sample freq for both files as a check
print sound4khz[0] # -> 4000
# print sound8khz[0] # -> 8000
```

```python
# setup a time scale for the x axis on plots that equates to the samaple counts of each file
def timeScale(sound):
#     np.ndarray(shape=sound[0],dtype=np.float32,
    return np.arange(0, sound[1].shape[0]/float(sound[0]), (1.0/sound[0]), np.float32)

time4khz = timeScale(sound4khz)
# time8khz = timeScale(sound8khz)
```

```python
# function to take in data and file name and output simple plotted graph files
def makegraph(sound, timeRange, sample_rate, title=''):
    plt.figure()
    plt.plot(timeRange, sound)
    plt.ylabel("Amplitude")
    plt.xlabel("Time (seconds)")
    plt.title(title)
    plt.suptitle("Sampled at {}Hz".format(sample_rate))
    #plt.savefig('/home/kak9699/Documents/code/SoundFiltering/' + sample_rate + '.png')
    #plt.show()
```

```python
# function to process audio data using a Hanning window and FFT
def FFTwithHann(audio):
    # gets the shape of the nparray passed in
    shape = audio.shape[0]
    # create and apply the Hanning Window to audio data
    window = hann(shape)
    audio = audio * window
    # process windowed audio data with FFT
    mags = abs(fft(audio))
    # convert to dB, if desired
    # mags = 20 * scipy.log10(mags)
    return mags
```

```python
# function to process audio data using a Gaussian Window and FFT
def FFTwithGaussian(audio, sigma):
    # gets the shape of the nparray passed in
    shape = audio.shape[0]
    # create and apply the Gaussian Window to audio data
    window = gaussian(shape, sigma)
    audio = audio * window
    # process windowed audio data with FFT
    mags = abs(fft(audio))
    # convert to dB, if desired
    # mags = 20 * scipy.log10(mags)
    return mags
```

```python
# plot graphs representing the raw freq data
makegraph(sound4khz[1], time4khz, sound4khz[0], 'Raw Import File Data')
# makegraph(sound8khz[1], time8khz, sound8khz[0], 'Raw Import File Data')
```

```python
# transform imported audio with Hanning Window and FFT 
Hann4khz = FFTwithHann(sound4khz[1])
# Hann8khz = FFTwithHann(sound8khz[1])

# plot FFT graphs
makegraph(Hann4khz, time4khz, sound4khz[0], 'FFT with Hanning Window')
# makegraph(Hann8khz, time8khz, sound8khz[0], 'FFT with Hanning Window')
```

```python
# transform imported audio with Gaussian Window and FFT 
Gauss4khz = FFTwithGaussian(sound4khz[1], 100)
# Gauss8khz = FFTwithGaussian(sound8khz[1])

# plot FFT graphs
makegraph(Gauss4khz, time4khz, sound4khz[0], 'FFT with Gaussian Window')
# makegraph(Gauss8khz, time8khz, sound8khz[0], 'FFT with Gaussian Window')
```
